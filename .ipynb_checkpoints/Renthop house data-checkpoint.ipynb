```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_train = pd.read_json('train.json')
data_test = pd.read_json('test.json')
```

```python
data = data_train.append(data_test)
```

```python
data = data.reindex(range(data.shape[0]))
```

```python
data['num_pic'] = data['photos'].apply(lambda x: len(x))
```

```python
#want only house with multiple bedrooms and high interest level
```

```python
listing_id = data.loc[(data['bedrooms']>1) & (data['interest_level'] == 'high') & (data['num_pic'] > 3)].sort_values(by = 'listing_id')
```

```python
import os
from os import listdir
from PIL import Image
from matplotlib.pyplot import imshow
import numpy as np

%matplotlib inline
```

```python
dir_image = "F:\\apartment_pic\\images"
```

```python
os.chdir(dir_image)
```

```python
def read_image(list_dir):
    n = 0
    while True:
        yield [ list_dir[n]+ "\\" +f for f in listdir(os.path.join(dir_image,list_dir[n]))]
        n += 1
```

```python
a = read_image(ld)
```

```python
for i in next(a):
    print(i)
    plt.figure()
    pil_im = Image.open(i, 'r')
    plt.imshow(np.asarray(pil_im))
```
